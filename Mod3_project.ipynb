```python
#groups crime by zip code and counts the occurence
crime_by_zip = crime['ZIP'].value_counts()
crime_by_zip
```

```python
crime_sorted_by_zip = crime.sort_values(by='ZIP')
```

```python
crime_sorted_by_zip
```

```python
#sort the data based on the crimes by crime category and zip code
sorted_d = crime.sort_values(by=['UCR CRIME CATEGORY', 'ZIP'])
```

```python
sorted_d
```

```python
#group crimes by zip and crime category, and counts occurence
crime_by_zip_and_category = crime.groupby(['ZIP', 'UCR CRIME CATEGORY']).size().unstack(fill_value=0)
crime_by_zip_and_category
```

```python
#sorts the data based on crime by category and premise type
type_c = crime.sort_values(by=['UCR CRIME CATEGORY','PREMISE TYPE'])
```

```python
type_c
```

```python
#group crimes by premise type and crime category, and counts occurence
crime_by_premise_and_category = crime.groupby(['PREMISE TYPE', 'UCR CRIME CATEGORY']).size().unstack(fill_value=0)
crime_by_premise_and_category
```
